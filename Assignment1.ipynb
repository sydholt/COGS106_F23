### Problem #1 
#### Make an array x containing 200 random numbers (samples) drawn from a normal distribution with a mean of 0.25 and a standard deviation of 1 and plot a histogram. Don't forget to set the random number seed.  



```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

mean = 0.25
std_dev = 1.0
num_samples = 200
x = np.random.normal(mean, std_dev, num_samples)

# Create a histogram
plt.hist(x, bins=20, edgecolor='k', alpha=0.65)
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Problem #2  
#### Find the mean, median, 25%, 75% and standard deviation of the array you created in Problem 1

```python
mean_x = np.mean(x)
median_x = np.median(x)
q1_x = np.percentile(x, 25)
q3_x = np.percentile(x, 75)
std_dev_x = np.std(x)

print("Mean:", mean_x)
print("Median:", median_x)
print("25th Percentile (Q1):", q1_x)
print("75th Percentile (Q3):", q3_x)
print("Standard Deviation:", std_dev_x)
```

### Problem #3 
#### Compute the cumulative sum of the array x (there is a numpy function that does htis) and store it in a variable w.  Make a plot of the number of samples versus w. 

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.cumsum(x)
sample_numbers = np.arange(1, len(w) + 1)

plt.plot(sample_numbers, w)
plt.title('Cumulative Sum of Array x')
plt.xlabel('Number of Samples')
plt.ylabel('Cumulative Sum (w)')

plt.show()
```

#### a. Calculate the mean reaction time for each subject in each condition.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("ReactionTimeData.csv")

means = data.groupby(['Subject', 'Condition'])['ReactionTime'].mean().unstack()

plt.figure(figsize=(8, 6))
conditions = ['Easy', 'Medium', 'Hard']

for condition in conditions:
    plt.hist(means[condition], bins=20, alpha=0.5, label=condition)

plt.title('Distribution of Mean Reaction Times Across Subjects')
plt.xlabel('Mean Reaction Time (ms)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

#c
p_25 = data.groupby(['Subject', 'Condition'])['ReactionTime'].quantile(.25).unstack()

plt.figure(figsize=(8, 6))
conditions = ['Easy', 'Medium', 'Hard']

for condition in conditions:
    plt.hist(p_25[condition], bins=20, alpha=0.5, label=condition)

plt.title('Distribution of 25th Percentile Reaction Times Across Subjects')
plt.xlabel('25% Reaction Time (ms)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

p_50 = data.groupby(['Subject', 'Condition'])['ReactionTime'].quantile(.5).unstack()

plt.figure(figsize=(8, 6))
conditions = ['Easy', 'Medium', 'Hard']

for condition in conditions:
    plt.hist(p_50[condition], bins=20, alpha=0.5, label=condition)

plt.title('Distribution of 50th Percentile Reaction Times Across Subjects')
plt.xlabel('50% Reaction Time (ms)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

p_75 = data.groupby(['Subject', 'Condition'])['ReactionTime'].quantile(.75).unstack()

plt.figure(figsize=(8, 6))
conditions = ['Easy', 'Medium', 'Hard']

for condition in conditions:
    plt.hist(p_75[condition], bins=20, alpha=0.5, label=condition)

plt.title('Distribution of 75th Percentile Reaction Times Across Subjects')
plt.xlabel('75% Reaction Time (ms)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

#d
# Need to figure out what happened to the last graph
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ReactionTimeData.csv")

data['Correct'] = data['ReactionTime'] == data['Correct']
accuracy_data = data.groupby(['Subject', 'Condition'])['Correct'].mean().unstack()

plt.figure(figsize=(10, 6))
conditions = ['Easy', 'Medium', 'Hard']

for condition in conditions:
    plt.bar(accuracy_data.index, accuracy_data[condition], label=condition, alpha=0.5)

plt.title('Accuracy in Each Condition for All Subjects')
plt.xlabel('Subject')
plt.ylabel('Accuracy')
plt.xticks(accuracy_data.index)
plt.legend()

plt.show()
```
